### 5.1.2 数组的操作
1. 访问数组
对数组里的元素进行操作，首先是要能够索引元素，即查询访问。  
索引：每个维度一个索引值，用逗号分隔。

```python
a = np.random.randint(2, 100, 24).reshape((3, 8))
a
```

```python
a[2, 6]  # 访问索引号为[2, 6]位置上的元素
```

```python
b =  a.reshape((2, 3, 4))  # 将a改为三维数组
b
```

```python
b[1, 2, 3]  # 访问索引号为[1, 2, 3]位置上的元素
```
